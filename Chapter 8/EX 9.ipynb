```python
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', as_frame=False, parser="auto")
```

#### Split it into a training set and a test set

```python
X_train, y_train = mnist.data[:60_000], mnist.target[:60_000]
X_test, y_test = mnist.data[60_000:], mnist.target[60_000:]
```

#### Train a random forest classifier on the dataset and time how long it takes

```python
rnd_clf = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
%time rnd_clf.fit(X_train, y_train)
```

#### Took 58.4s to run the RandomForestClassifier  model, now lets evaluate on the test set

```python
from sklearn.metrics import accuracy_score

y_pred = rnd_clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

#### Using PCA to reduce the dataset dimensionality wijt explained variance ration of 95%

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95)
X_train_reduced = pca.fit_transform(X_train)
```

#### Training a new RandomForestClassifier and timing how long it takes

```python
rnd_clf_with_pca = RandomForestClassifier(n_estimators=100, random_state=42)
%time rnd_clf_with_pca.fit(X_train_reduced, y_train)
```

#### Wow  about 2.9x slower to train than the initial dataset, now lets evaluate the classifier on the test set

```python
X_test_reduced = pca.transform(X_test)

y_pred = rnd_clf_with_pca.predict(X_test_reduced)
accuracy_score(y_test, y_pred)
```

####  it also slightly reduce the  performance 

#### Now trying with an SGDClassifier and also timing it 

```python
from sklearn.linear_model import SGDClassifier

sgd_clf = SGDClassifier(random_state=42)
%time sgd_clf.fit(X_train, y_train)
```

```python
y_pred = sgd_clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
sgd_clf_with_pca = SGDClassifier(random_state=42)
%time sgd_clf_with_pca.fit(X_train_reduced, y_train)
```

```python
y_pred = sgd_clf_with_pca.predict(X_test_reduced)
accuracy_score(y_test, y_pred)
```

#### In summary reducing the dimensioality gave us about 5.4x speed up and slightly improved the performance
